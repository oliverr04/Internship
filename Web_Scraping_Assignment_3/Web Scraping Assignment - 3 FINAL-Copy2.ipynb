```python
import selenium
from selenium import webdriver
import pandas as pd


from bs4 import BeautifulSoup
import requests

import time

from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait

import warnings
warnings.filterwarnings("ignore")
```

```python
driver_10=webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe") 
time.sleep(5)
```

```python
driver_10.get('https://www.hostelworld.com/')
driver_10.maximize_window()
```

```python
search_loc = driver_10.find_element_by_id('search-input-field')
search_loc.send_keys("London")
time.sleep(5)
```

```python
london = driver_10.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]/div')
london.click()
time.sleep(5)
```

```python
search_btn10 = driver_10.find_element_by_id('search-button')
search_btn10.click()
```

```python
url10="https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2022-05-16&to=2022-05-19&guests=2&page=1"
```

```python
driver_10.get(url10)
```

```python
HName = driver_10.find_elements_by_xpath("//h2[@class='title title-6']/a")
Hostel_Names = []
for i in HName:
    Hostel_Names.append(i.text.replace('\n',''))
    
Hostel_Names
```

```python
Dist = driver_10.find_elements_by_xpath("//span[@class='description']")
Distance = []
for i in Dist:
    Distance.append(i.text.replace('\n',''))
    
Distance
```

```python
rating = driver_10.find_elements_by_xpath("//div[@class='score orange big']")
Ratings = []
for i  in rating:
    try:
        Ratings.append(i.text)
    except NoSuchElementException as e:
        Ratings.append("-")
```

```python
Ratings
```

```python
T_Reviews = driver_10.find_elements_by_xpath("//div[@class='reviews']")
Total_Reviews =[]
for i in T_Reviews:
    Total_Reviews.append(i.text)
    
Total_Reviews
```

```python
O_reviews = driver_10.find_elements_by_xpath("//div[@class='keyword']")[0:30]
Overall_Reviews =[]
for i in O_reviews:
    Overall_Reviews.append(i.text)
    
Overall_Reviews
```

```python
privates = driver_10.find_elements_by_xpath("//div[@class='price title-5']")[0:30]
Privates_from_price1 =[]
for i in privates:
    try:
        Privates_from_price1.append(i.text)
    except NoSuchElementException as e:
        Privates_from_price1.append("-")
    
    
Privates_from_price1
```

```python
Privates_from_price1 = Privates_from_price1[0::2]
```

```python
Privates_from_price1
```

```python
dorms = driver_10.find_elements_by_xpath("//div[@class='price title-5']")
Dorms_from_price =[]
for i in dorms:
    try:
        Dorms_from_price.append(i.text)
    except NoSuchElementException as e:
        Dorms_from_price.append("-")
        
Dorms_from_price
```

```python
Dorms_from_price = Dorms_from_price[1::2]
Dorms_from_price
```

```python
facil = driver_10.find_elements_by_xpath("//div[@class='facilities-label facilities']")
Facilities=[]
for i in facil:
    Facilities.append(i.text.replace('\n',''))
    
Facilities
```

```python
prop_desc = driver_10.find_elements_by_xpath("//div[@class='rating-factors prop-card-tablet rating-factors small']")
Property_Description =[]
for i in prop_desc:
    Property_Description.append(i.text.replace('\n',' '))
    
Property_Description
```

```python
len(Hostel_Names),len(Distance),len(Ratings),len(Total_Reviews),len(Overall_Reviews),len(Privates_from_price1),len(Dorms_from_price),len(Facilities),len(Property_Description)
```

```python
import pandas as pd
Hostels=pd.DataFrame({})
Hostels['Hostel Name']= Hostel_Names[0:15]
Hostels['Distance from City Centre']= Distance[0:15]
Hostels['Ratings']= Ratings[0:15]
Hostels['Total Reviews']= Total_Reviews[0:15]
Hostels['Overall Reviews']= Overall_Reviews[0:15]
Hostels['Privates from Price']= Privates_from_price1[0:15]
Hostels['Dorms from Prices']= Dorms_from_price[0:15]
Hostels['Property Description']= Property_Description[0:15]
Hostels['Facilities']= Facilities[0:15]
```
