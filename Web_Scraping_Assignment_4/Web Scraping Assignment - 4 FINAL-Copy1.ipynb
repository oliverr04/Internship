```python
dataset_url = driver10.find_elements_by_xpath("//p[@class='normal']//b/a")    
urls10 = []     
for i in dataset_url:
    urls10.append(i.get_attribute("href"))
```

```python
Dataset_name = []
Dataset_type = []
Task = []
Attr_type = []
Instances = []
n_attributes = []
Year = []
```

```python
for i in urls10:
    driver10.get(i)
    time.sleep(3)
    
    # Dataset Name
    try: 
        ds_name = driver10.find_element_by_xpath("//span[@class='heading']")
        Dataset_name.append(ds_name.text)
    except NoSuchElementException:
        Dataset_name.append('-')
    time.sleep(1)
    
    # Dataset Type
    try:
        ds_type = driver10.find_element_by_xpath("//table[@border='1']//tbody/tr/td[2]")
        if ds_type.text == "N/A": raise NoSuchElementException
        Dataset_type.append(ds_type.text)
    except NoSuchElementException:
        Dataset_type.append('-')
    time.sleep(1)
    
    # Attribute Type
    try:
        atype = driver10.find_element_by_xpath("//table[@border='1']//tbody/tr[2]/td[2]")
        if atype.text == "N/A": raise NoSuchElementException
        Attr_type.append(atype.text)
    except NoSuchElementException:
        Attr_type.append('-')
    time.sleep(1)
    
    # Task
    try:
        task = driver10.find_element_by_xpath("//table[@border='1']//tbody/tr[3]/td[2]")
        if task.text == "N/A": raise NoSuchElementException
        Task.append(task.text)
    except NoSuchElementException:
        Task.append('-')
    time.sleep(1)
    
    # No of instances
    try:
        inst = driver10.find_element_by_xpath("//table[@border='1']//tbody/tr/td[4]")
        if inst.text == "N/A": raise NoSuchElementException
        Instances.append(inst.text)
    except NoSuchElementException:
        Instances.append('-')
    time.sleep(1)
    
    # No of attributes
    try:
        attr = driver10.find_element_by_xpath("//table[@border='1']//tbody/tr[2]/td[4]")
        if attr.text == "N/A": raise NoSuchElementException
        n_attributes.append(attr.text)
    except NoSuchElementException:
        n_attributes.append('-')
    time.sleep(1)
    
    # Year
    try:
        year = driver10.find_element_by_xpath("//table[@border='1']//tbody/tr[2]/td[6]")
        if year.text == "N/A": raise NoSuchElementException
        Year.append(year.text[:4])
    except NoSuchElementException:
        Year.append('-')
    time.sleep(1)
```

```python
len(Dataset_name),len(Dataset_type),len(Task),len(Attr_type),len(Instances),len(n_attributes),len(Year)
```

```python
UCI=pd.DataFrame({})
UCI['Dataset Name'] = Dataset_name
UCI['Dataset Type'] = Dataset_type
UCI['Task'] = Task
UCI['Attribute type'] = Attr_type
UCI['No of instance'] = Instances
UCI['No of attributes'] = n_attributes
UCI['Year'] = Year
```
